# 載入資料集

```python
# 引入所需的庫
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# 使用 sklearn 的 load_iris 函數載入 iris 資料集
iris = load_iris()

# 將 iris 資料集轉換為 DataFrame 格式，方便後續的資料處理和分析
# np.c_ 是 numpy 的一個函數，用於將 iris['data'] 和 iris['target'] 兩個陣列進行串接
# columns 參數用於指定 DataFrame 的欄位名稱
df_data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species'])

# 顯示 DataFrame
df_data
```

## 直方圖

```python
#直方圖 histograms

# 使用 pandas DataFrame 的 hist 函數繪製直方圖
# alpha 參數設定透明度，範圍為 0（完全透明）到 1（完全不透明）
# layout 參數設定子圖的排列方式，(3,3) 表示 3x3 的格子
# figsize 參數設定圖形的大小，(12, 8) 表示寬度為 12，高度為 8
# bins 參數設定直方圖的區間數量
df_data.hist(alpha=0.6, layout=(3,3), figsize=(12, 8), bins=10)

# 使用 tight_layout 函數自動調整子圖參數，使之填充整個圖形區域
plt.tight_layout()

# 使用 show 函數顯示圖形
plt.show()

```
